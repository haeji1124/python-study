# 클로저


클로저를 사용하면 프로그램의 흐름을(환경) 변수에 저장할 수 있다. 원할때 저장된 흐름(환경)을 다시 꺼내서 사용할 수 있다.

클로저는 지역 변수와 코드를 묶어서 사용하고 싶을 때 활용한다. 

클로저에 속한 지역 변수는 바깥에서 직접 접근할 수 없으므로 데이터를 숨기고 싶을 때 활용한다.


하지만, 파이썬에서는 클래스를 지원하기 때문에 클로저의 효용가치가 많이 떨어진다.

왜냐하면 굳이 클로저로 구현하지 않고 클래스를 만들어서 클래스 내부에 함수를 정의해 놓으면 

언제든지 객체를 만들어서 해당 인스턴스를 불러올 수 있기 떄문이다.

따라서 클로저 같은 경우 클래스를 지원하지 않는 언어에서 클래스를 재현하기 위해 사용할 수 있다. 그러나 함수가 몇개 없는 미니 클래스라면 클로저를 활용하는 것도 좋다.

return 함수 ---> 함수의 주소를 리턴함으로 outter 함수 호출하여 inner함수를 계속 쓸 수 있다. 주소저장되어 있어서

````python
########################################################
# 문제 1
########################################################
# n = int(input())
n = 6

print("별1개")
for _ in range(n):
    print("*")  # oneStar

print("-"*50)
print("왼쪽정렬 피라미드")
# 왼쪽정렬 피라미드
for i, _ in enumerate(range(n)):
    print("*"*(i+1))

print("-"*50)
print("오른쪽정렬 피라미드")
# 오른쪽정렬 피라미드
for i, _ in enumerate(range(n)):
    print(" "*(n-i-1)+"*"*(i+1))
print("-"*50)
print("왼쪽정렬 피라미드 거꾸로")
# 왼쪽정렬 피라미드 거꾸로
for i, _ in enumerate(range(n)):
    print("*"*(n-i))

print("-"*50)
print("오른쪽정렬 피라미드 거꾸로")
# 오른쪽정렬 피라미드 거꾸로
for i, _ in enumerate(range(n)):
    print(" "*(i)+"*"*(n-i))


print("-"*50)
print("별4개")
# 별4개
for i, _ in enumerate(range(n)):
    print("*"*4)
print("-"*50)
print("가운데 정렬 피라미드")
# 가운데 정렬 피라미드
for i, _ in enumerate(range(n)):
    print(" "*(n-i-1)+"*"*(i*2+1))

print("-"*50)
print("가운데 정렬 피라미드 거꾸로")
# 가운데 정렬 피라미드
for i, _ in enumerate(range(n)):
    print(" "*(i)+"*"*((n-i-1)*2+1))
````

어쨌든.... 즉 함수 밖에서도 함수 내에 변수 i를 조절할 수 있는 방법이 필요한 상황...으로 이럴때 클로져를 사용 해야한다고 생각합니다.

또한 이미 구현된 함수에 대해서는 수정하고 싶지 않지만 뭔가 변화가 필요한 경우에도 클로저를 사용해야한다고 생각합니다..?

별 피라미드 함수가 이미 존재하는지만, 내부는 수정하지 않고 함수를 확장하고 싶을 때...예를들면 피라미드 함수 수행시간을 알고 싶거나, 함수 실행을 지연시킵니다.

````python
### 별 피라미드 함수 Creator : eval 없이###
def create_pyramid(func_span):
    def inner(num):
        for i in range(num):
            num_indent, num_stars = func_span(num, i)
            print(f"{' ' * num_indent}{'*' * num_stars}")

    return inner

### 함수 생성 ###
line = create_pyramid(lambda n, i: (0, 1))
rect = create_pyramid(lambda n, i: (0, 4))
left = create_pyramid(lambda n, i: (0, i+1))
right = create_pyramid(lambda n, i: ((n-1)-i, i+1))
center = create_pyramid(lambda n, i: ((n-1)-i, 2*i+1))
rev_left = create_pyramid(lambda n, i: (0, n-i))
rev_right = create_pyramid(lambda n, i: (i, n-i))
rev_center = create_pyramid(lambda n, i: (i, 2*((n-1)-i)+1))


### 테스트 ###
num = 5
for pyramid in [line, rect, left, right, center, rev_left, rev_right, rev_center]:
    pyramid(num)
    print()



### 추가 동작 ###
def get_span(n, y, equation):       # span = (space개수, star개수)
    # n x n 영역에서, equation을 사용하여 y위치에 대한 x위치를 구함
    for x in range(n):
        if equation(x, y):
            return (x, n - 2*x)     # x는 space 개수로 활용. span길이는 width - 2*(공백길이) = n-2*x
    return n//2                     # 미사용

diamond = create_pyramid(lambda n, i: get_span(n, i, lambda x, y: abs(x*2/(n-1)-1) + abs(y*2/(n-1)-1) <= 1))
circle = create_pyramid(lambda n, i: get_span(n, i, lambda x, y: (((x*2/(n-1))-1)**2 + (y*2/(n-1)-1)**2) <= 1))

num = 17
for pyramid in [diamond, circle]:
    pyramid(num)
    print()
````

````python
def create_pyramid_delta(pos_deltas):
    def inner(num):
        (pos0, delta0), (pos1, delta1) = pos_deltas(num)
        for i in range(num):
            indent, stars = (' ' * pos0), ('*' * (pos1 - pos0 + 1))
            pos0, pos1 = (pos0 + delta0), (pos1 + delta1)
            print(f'{indent}{stars}')

    return inner

line = create_pyramid_delta(lambda n: ((0, 0), (0, 0)))
rect = create_pyramid_delta(lambda n: ((0, 0), (3, 0)))
left = create_pyramid_delta(lambda n: ((0, 0), (0, 1)))
right = create_pyramid_delta(lambda n: ((n-1, -1), (n-1, 0)))
center = create_pyramid_delta(lambda n: ((n-1, -1), (n-1, 1)))
rev_left = create_pyramid_delta(lambda n: ((0, 0), (n-1, -1)))
rev_right = create_pyramid_delta(lambda n: ((0, 1), (n-1, 0)))
rev_center = create_pyramid_delta(lambda n: ((0, 1), (2*(n-1), -1)))


num = 7
for pyramid in [line, rect, left, right, center, rev_left, rev_right, rev_center]:
    pyramid(num)
    print()
````

````python
def create_pyramid_equation(equation):
    mapper = lambda coord: '*' if equation(*coord) else ' '

    def inner(num):
        for y in range(num):    # (x, y): num x num region
            line = ''.join(map(mapper, [(num, x, y) for x in range(num)]))
            print(line.rstrip())
    return inner


line = create_pyramid_equation(lambda n, x, y: x == 0)
left = create_pyramid_equation(lambda n, x, y: x <= y)
right = create_pyramid_equation(lambda n, x, y: (x+y) >= n-1)
rev_right = create_pyramid_equation(lambda n, x, y: x >= y)
rev_left = create_pyramid_equation(lambda n, x, y: (x+y) <= n-1)
rect = create_pyramid_equation(lambda n, x, y: x < 4)
diamond = create_pyramid_equation(lambda n, x, y: abs(x*2/(n-1)-1) + abs(y*2/(n-1)-1) <= 1)
circle = create_pyramid_equation(lambda n, x, y: (((x*2/(n-1))-1)**2 + (y*2/(n-1)-1)**2) <= 1)


num = 15
for pyramid in [line, left, right, rev_left, rev_right, rect, diamond, circle]:
    pyramid(num)
    print()
````

````python
def create_edge_pyramid(func_span):
    def inner(height):
        for i in range(height):
            num_indent, num_stars = func_span(height, i)
            indent = ' ' * num_indent
            if i == 0 or i == (height - 1) or num_stars < 2:    # the first/last line or 줄수가 충분하지 않을 경우
                line = '*' * num_stars
            else:
                line = '*' + ' ' * (num_stars-2) + '*'          # 양 끝을 '*'로 출력하고 그 사이는 ' '
            print(f'{indent}{line}')

    return inner

def get_span(n, y, equation):       # span = (space개수, star개수)
    # n x n 영역에서, equation을 사용하여 y위치에 대한 x위치를 구함
    for x in range(n):
        if equation(x, y):
            return (x, n - 2*x)     # x는 space 개수로 활용. span길이는 width - 2*(공백길이) = n-2*x
    return n//2                     # 미사용

line = create_edge_pyramid(lambda n, i: (0, 1))
rect = create_edge_pyramid(lambda n, i: (0, 4))
left = create_edge_pyramid(lambda n, i: (0, i+1))
right = create_edge_pyramid(lambda n, i: ((n-1)-i, i+1))
center = create_edge_pyramid(lambda n, i: ((n-1)-i, 2*i+1))
rev_left = create_edge_pyramid(lambda n, i: (0, n-i))
rev_right = create_edge_pyramid(lambda n, i: (i, n-i))
rev_center = create_edge_pyramid(lambda n, i: (i, 2*((n-1)-i)+1))
diamond = create_edge_pyramid(lambda n, i: get_span(n, i, lambda x, y: abs(x*2/(n-1)-1) + abs(y*2/(n-1)-1) <= 1))
circle = create_edge_pyramid(lambda n, i: get_span(n, i, lambda x, y: (((x*2/(n-1))-1)**2 + (y*2/(n-1)-1)**2) <= 1))

num = 15
for pyramid in [line, rect, left, right, center, rev_left, rev_right, rev_center, diamond, circle]:
    pyramid(num)
    print()
````
